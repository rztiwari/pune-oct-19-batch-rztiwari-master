```python
#Import pandas libraries
import pandas as pd 
from pandas import Series
import numpy as np
#Read the CSV file
Laliga = pd.read_csv('Laliga.csv')
Laliga.head()
```

```python
#Clean and format the Dataframe. The first row in the csv file was empty.
#Get rid of the empty row and set the axis correctly to read the DF.
Laliga.set_axis(Laliga[:1].values[0], axis=1, inplace=True)
Laliga.drop([0], axis=0, inplace=True)
Laliga
```

```python
#Step1: Replace dashes with 0 to make sure you can perform arithmetic operations on the data
Laliga.replace('-', int(0), True)
Laliga
```

```python
#Step 2: Print all the teams which have started playing between 1930-1980. use “Since/lastApp” column
index = 0
indexToStore = []
for data in Laliga['Since/LastApp']:
    #print(data[0:4])
    if(int(data[0:4]) >= 1930 and int(data[0:4]) <= 1980):
        indexToStore.append(index)
    index = index + 1
indexToStore
TeamsStartedBtw1930AND1980 = Laliga.iloc[indexToStore, :]
TeamsStartedBtw1930AND1980
```

```python
#Step 3: Print the list of teams which came Top 5 in terms of points
Laliga['Points'] = pd.to_numeric(Laliga['Points'])
Laliga['GoalsFor'] = pd.to_numeric(Laliga['GoalsFor'])
Laliga['GoalsAgainst'] = pd.to_numeric(Laliga['GoalsAgainst'])
Laliga['GamesPlayed'] = pd.to_numeric(Laliga['GamesPlayed'])
Laliga['GamesWon'] = pd.to_numeric(Laliga['GamesWon'])
Laliga['BestPosition'] = pd.to_numeric(Laliga['BestPosition'])
Top5 = Laliga.sort_values(by='Points', ascending=False).head()
Top5
```

```python
def Goal_diff_count(Teams):
    return Series((Teams['GoalsFor'] - Teams['GoalsAgainst']).values, Teams['Team'])
Goal_diff_count(Laliga)
#Sort the data
SortedData = Goal_diff_count(Laliga).sort_values()
print(SortedData)
TeamMinGoalDiff = SortedData.index[0]
print('Team with min goal difference' , TeamMinGoalDiff)
TeamMaxGoalDiff = SortedData.index[SortedData.size -1]
print('Team with max goal difference', TeamMaxGoalDiff)
```

```python
#Create a new column with the name “Winning Percent” and append it to the data set
Laliga['Winning Percent'] = Laliga['GamesWon']/Laliga['GamesPlayed']*100
#If there are any numerical error, replace it with 0%
Laliga.fillna(0)
#Print the top 5 teams which have the highest Winning percentage
Laliga.sort_values(by='Winning Percent', ascending=False).head()
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions
Laliga.groupby('BestPosition').sum().sort_values(by='BestPosition').iloc[:, :1]
```
